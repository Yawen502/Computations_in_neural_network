```python
#plot
import matplotlib.pyplot as plt
import matplotlib
# extend multiscle to length 500 with the last value

#plot a vertical line at end of multiscale

multiscale = np.append(multiscale0, np.repeat(multiscale0[-1], 500 - len(multiscale0)))
lstm = np.append(lstm0, np.repeat(lstm0[-1], 500 - len(lstm0)))
vanilla = np.append(vanilla0, np.repeat(vanilla0[-1], 500 - len(vanilla0)))
dale = np.append(dale0, np.repeat(dale0[-1], 500 - len(dale0)))
multiscale_dale = np.append(multiscale_dale0, np.repeat(multiscale_dale0[-1], 500 - len(multiscale_dale0)))
cbgru = np.append(cbgru0, np.repeat(cbgru0[-1], 500 - len(cbgru0)))
plt.plot(multiscale)
plt.plot(lstm)
plt.plot(vanilla)
plt.plot(dale)
plt.plot(multiscale_dale)
plt.plot(cbgru)
# set y limit
plt.ylim(0, 3)
plt.legend(['Multiscale', 'LSTM', 'Vanilla', 'Dale', 'Multiscale_Dale', 'CBGRU'])
plt.title('MSE Error')
plt.show()
```
