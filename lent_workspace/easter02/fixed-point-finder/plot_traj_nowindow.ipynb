### Model Explanation

Multiscale-var:\
$v_{t+1} = (1-z) v_t + z(Wf(v_t)+Px_t +b_z)$

CB-GRU-var:\
z becomes a constant that z proportional to $f(b_z)$

CB-GRU-var2:\
Change voltage equation to firing rate equation (same as multiscale) to check consistency. It should be the same as simple GRU.

CB-GRU-var3:\
z becomes a constant that z proportional to $f(b_z)$. Change voltage equation to firing rate equation (same as multiscale) to check consistency.
